```python
import math
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
L=6.28
nx=10001
dx=6.28/nx
x=numpy.linspace(0,6.28,num=nx)
u=numpy.zeros(nx)
```

```python
u1=numpy.cos(x)/((x+1)**4)-4*numpy.sin(x)/((x+1)**5)
```

```python
u2=(2*(numpy.sin(x+dx)/((x+dx+1)**4))+3*(numpy.sin(x)/((x+1)**4))
-6*(numpy.sin(x-dx)/((x-dx+1)**4))+(numpy.sin(x-2*dx)/((x-2*dx+1)**4)))/(6*dx)
u=u1-u2
```

```python
pyplot.rcParams['font.family']='serif'
pyplot.rcParams['font.size']=16
pyplot.figure(figsize=(6.0,2.0))
pyplot.title('exact solution')
pyplot.xlabel('x')
pyplot.ylabel('u1')
pyplot.grid()
pyplot.plot(x, u1, color='C1', linestyle='-', linewidth=1)
pyplot.xlim(0.0, L)
pyplot.ylim(-0.15, 1);

pyplot.rcParams['font.family']='serif'
pyplot.rcParams['font.size']=16
pyplot.figure(figsize=(6.0,2.0))
pyplot.title('approximate solution')
pyplot.xlabel('x')
pyplot.ylabel('u2')
pyplot.grid()
pyplot.plot(x, u2, color='C0', linestyle='-', linewidth=1)
pyplot.xlim(0.0, L)
pyplot.ylim(-0.15, 1);

pyplot.rcParams['font.family']='serif'
pyplot.rcParams['font.size']=16
pyplot.figure(figsize=(6.0,2.0))
pyplot.title('u1-u2')
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u, color='C1', linestyle='-', linewidth=1)
pyplot.xlim(0.0, L)
pyplot.ylim(-0.00000000001,0.00000000001);
```
